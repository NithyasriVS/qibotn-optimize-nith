```python
import argparse
from timeit import default_timer as timer

import quimb as qu
from qiskit import QuantumCircuit

from qibo import __version__, gates, tests
from qibo.models import Circuit, QFT
from qibo.tests.utils import random_state

print('quimb version:', qu.__version__)
```

```python
num_qubits = 6

print('Testing for {} qubits'.format(num_qubits))

# generate the initial state
initial_state = random_state(num_qubits)

# get the qibo circuit, convert it to a qasm circuit
start = timer()
qibo_qft_circuit = QFT(num_qubits)
final_state = qibo_qft_circuit(initial_state)
end = timer()

# print(qibo_qft_circuit.summary())
print('qibo QFT circuit:')
print(qibo_qft_circuit.draw())
# print(final_state.state())
# print("qibo execution time is " + str(end - start))

# convert the circuit format from qibo to qasm
qasm_qft_circuit = qibo_qft_circuit.to_qasm()

# convert the circuit format from qasm to qiskit
qiskit_qft_circuit = QuantumCircuit.from_qasm_str(qasm_qft_circuit)
print('\nqiskit QFT circuit:')
qiskit_qft_circuit.draw()
```
